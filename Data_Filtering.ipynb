```python
import pandas as pd
import numpy as np
```

```python
data ={'string_col' : ['1','2', '3','4'],
        'int_float' : [1, 2, 3.2, 4.6],
        'mix_col' : ['a', '-2', 3, 4],
        'missing' :[1,2,3,np.nan],
        'money' : ['$1,500.00', '$3,436.99', '$958.99','$1,198.00']
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
#filtering out data that ha only one certain x-tics
df['int_float'] == 2
```

```python
filt = (df['int_float'] == 2)
```

```python
df[filt]
```

```python
#option 2
df.loc[filt]
```

```python
#if you want filterimng with an extra conditions also
df.loc[filt, 'string_col']
```

```python
# & = and, | = or
filt = (df['int_float'] == '2') & (df['mix_col'] == '-2')
```

```python
df.loc[filt]
```

```python
filt = (df['int_float'] == '2') | (df['mix_col'] == '-2')
```

```python
df.loc[filt]
```

```python
#this works with large data but since we havent loaded might practice on large data just get the skils
high_ranks = (df['int_float'] > 2)
```

```python
df.loc[high_ranks]
```

```python
#when we want to have extra conditions
df.loc[high_ranks, ['money', 'string_col']]
```

```python
#if yiu want to filt out some x-tics in a certain column
numbers =  ['1','2']
filt = df['string_col'].isin(numbers)
```

```python
df.loc[filt, 'string_col']
```
